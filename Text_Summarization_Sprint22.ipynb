```python
#MODULE 6

#NEW iteration of Summary models
import pandas as pd

#LexRank uses Cosine similarity between two sentences, to build a weighted graph for all the sentences in the document.
#it also takes care of the chosen top most sentences will not too similar to each other.
from sumy.summarizers.lex_rank import LexRankSummarizer

#LSA (Latent Semantic Analysis) is based on Singular value decomposition(SVD). It reduces the data into lower-dimensional space.
#LSA performs spatial decomposition and captures information in a singular vector and the magnitude of so singular vector will represent the importance.
#it has the capability to extract semantically related sentences but its computation is complex.
from sumy.summarizers.lsa import LsaSummarizer

# ... (rest of  imports, cleaning function, etc.)
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
import re
from gensim.utils import simple_preprocess
from nltk.stem import WordNetLemmatizer

# Define an iterative summarization loop to be called later (depends on the summary function defined below it.)
def generate_summaries_for_all_comments(df):
    summary_df = pd.DataFrame(columns=["doc_id","Text", "Topic/Subtopic", "Summarizer_type", "Sentence_count", "Summary","Human_Summary"])
    for index, row in df.iterrows():
        text = row['Text']
        comment = row["Topic/Subtopic"]
        Human_Summary=row["Human_Summary"]
        for summarizer_type in ["lexrank"]:    # prior: ["lexrank", "lsa"]
            for sentence_count in [3,4]:  # Adjust sentence counts as needed
                summary = generate_summary(text, summarizer_type, sentence_count)
                new_row = {
                    'doc_id': doc_id,
                    "Text": text,
                    "Topic/Subtopic": comment,
                    "Summarizer_type": summarizer_type,
                    "Sentence_count": sentence_count,
                    "Summary": summary,
                    "Human_Summary": Human_Summary                }
                summary_df = summary_df.append(new_row, ignore_index=True)

    return summary_df

# Define a function for flexible summarization function
def generate_summary(text, summarizer_type, sentence_count):
    parser = PlaintextParser.from_string(text, Tokenizer("english"))
    if summarizer_type == "lexrank":
        summarizer = LexRankSummarizer()
    elif summarizer_type == "lsa":  # Added for LSA option
        summarizer = LsaSummarizer()
    else:
        raise ValueError("Invalid summarizer type. Please choose 'lexrank' or 'lsa'.")
    summary_vector = summarizer(parser.document, sentences_count=sentence_count)
    return " ".join([str(sentence) for sentence in summary_vector])

# Create summary_df by applying the summary functions defined above.
summary_df = generate_summaries_for_all_comments(summary_df_EachTopic_human)

# Save the DataFrame to an Excel file
summary_df.to_excel('summary_output.xlsx', index=False)

summary_df.head()
```

```python
#MODULE 7

#NEW Question Answering Script
import pandas as pd
from transformers import pipeline
from docx import Document

#Use ask_llm function definition from above.
def ask_llm(prompt, text):
    # Send the request to the LLM
    response = client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content": f"{prompt} {text}",
            }
        ],
        model="mixtral-8x7b-32768",
    )

    # Return the response from the LLM
    return response.choices[0].message.content


def process_text_and_comments(df):
    answers_df = pd.DataFrame(columns=["doc_id","Text", "Topic/Subtopic", "Derived_Topic","Human_Summary"])

    for index, row in df.iterrows():
      text = row['Text']
      comment = row['Topic/Subtopic']  # Access column using backticks
      Human_Summary =row["Human_Summary"]
      try:
            # Answer questions using the LLM
            topic_answer = ask_llm("Write a 100-word summary of the context provided.", row['Text'])
            #concerns_answer = qa_pipeline({"question": "What are the author's most important concerns? (list up to 3 concerns)", "context": text})
            #solutions_answer = qa_pipeline({"question": "What solutions if any does the author propose?", "context": text})

            # Extract answers from LLM output
            Derived_Topic = topic_answer["answer"]
            #concerns = concerns_answer["answer"]
            #solutions = solutions_answer["answer"]

            # Add answers to the DataFrame using the actual comment value
            new_row = {
                "doc_id": doc_id,
                "Text": text,
                "Topic/Subtopic": comment,  # Use the actual comment value
                "Derived_Topic": Derived_Topic,
                #"Concerns": concerns,
                #"Solutions": solutions,
                "Human_Summary": Human_Summary
            }
            answers_df = answers_df.append(new_row, ignore_index=True)
      except Exception as e:
            print(f"Error processing text {index}: {text}\n{e}")

    return answers_df

#Use function defined above, to generate answers for each question to each,
# comment, then save to 'answers_df':
answers_df = process_text_and_comments(summary_df_EachTopic_human)
answers_df.to_excel('document_answers.xlsx', index=False)

print(answers_df)
```

```python
#MODULE 6 ALTERNATE FORM - USE PARAPHRASED SUMMARY

import pandas as pd
from sumy.summarizers.lex_rank import LexRankSummarizer
from sumy.summarizers.lsa import LsaSummarizer
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from nltk.corpus import wordnet
from nltk.tag import pos_tag
from nltk.tokenize import word_tokenize
nltk.download('wordnet')
# ... (rest of  imports, cleaning function, etc.)
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
import re
from gensim.utils import simple_preprocess
from nltk.stem import WordNetLemmatizer

# Define an iterative summarization loop to be called later (depends on the summary function defined below it.)
# Define an iterative summarization loop to be called later
def generate_summaries_for_all_comments_para(df):
    summary_df_paraphrased = pd.DataFrame(columns=["doc_id", "Text", "Topic/Subtopic", "Summarizer_type", "Sentence_count", "Summary", "Paraphrased_Summary", "Human_Summary"])  # Ensure "Paraphrased_Summary" is included here

    # Add the rule-based paraphraser
    for index, row in df.iterrows():
        text = row['Text']
        comment = row["Topic/Subtopic"]
        Human_Summary = row["Human_Summary"]
        for summarizer_type in ["lexrank", "lsa"]:
            for sentence_count in [7,8,9]:
                summary = generate_summary(text, summarizer_type, sentence_count)
                paraphrased_summary = rule_based_paraphraser(summary)
                new_row = {
                    'doc_id': doc_id,
                    "Text": text,
                    "Topic/Subtopic": comment,
                    "Summarizer_type": summarizer_type,
                    "Sentence_count": sentence_count,
                    "Summary": summary,
                    "Paraphrased_Summary": paraphrased_summary,  # Add the paraphrased_summary here
                    "Human_Summary": Human_Summary
                }
                summary_df_paraphrased = summary_df_paraphrased.append(new_row, ignore_index=True)  # Append to the correct DataFrame

    return summary_df_paraphrased

# Define a function for flexible summarization function
def generate_summary(text, summarizer_type, sentence_count):
    parser = PlaintextParser.from_string(text, Tokenizer("english"))
    if summarizer_type == "lexrank":
        summarizer = LexRankSummarizer()
    elif summarizer_type == "lsa":  # Added for LSA option
        summarizer = LsaSummarizer()
    else:
        raise ValueError("Invalid summarizer type. Please choose 'lexrank' or 'lsa'.")
    summary_vector = summarizer(parser.document, sentences_count=sentence_count)
    return " ".join([str(sentence) for sentence in summary_vector])

# Define a simple rule-based paraphraser
def rule_based_paraphraser(text):
    tokens = word_tokenize(text)
    pos_tags = pos_tag(tokens)

    paraphrased_tokens = []

    for word, pos in pos_tags:
        if pos.startswith('V'):  # Verb
            synonyms = [synset.lemmas()[0].name() for synset in wordnet.synsets(word) if synset.lemmas()]
            if synonyms:
                paraphrased_tokens.append(synonyms[0])
            else:
                paraphrased_tokens.append(word)
        else:
            paraphrased_tokens.append(word)

    paraphrased_text = ' '.join(paraphrased_tokens)
    return paraphrased_text

# Create summary_df by applying the summary functions defined above.
summary_df_paraphrased = generate_summaries_for_all_comments_para(summary_df_EachTopic_human)

# Save the DataFrame to an Excel file
summary_df_paraphrased.to_excel('summary_output_paraphrased.xlsx', index=False)

summary_df_paraphrased.head()
```
